**Load Data**

```python
import pandas as pd
```

```python
df = pd.read_csv("loan_stage1_cleaned.csv")
```

**Convert Categorical Codes**

```python
df['loan_limit'] = df['loan_limit'].replace({
    'cf':'conforming',
    'ncf':'non_conforming'
})


df['approv_in_adv'] = df['approv_in_adv'].replace({
    'nopre':'no_preapproval',
    'pre':'preapproved'
})


df['neg_ammortization'] = df['neg_ammortization'].replace({
    'not_neg':'no',
    'neg_amm':'yes'
})


df['interest_only'] = df['interest_only'].replace({
    'not_int':'no',
    'int_only':'yes'
})


df['business_or_commercial'] = df['business_or_commercial'].replace({
    'b/c':'business',
    'nob/c':'non_business'
})
```

**Handle Outliers**

```python
important_cols = ['loan_amount','income','credit_score','ltv']

for col in important_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]
```

**Feature Engineering**

```python
df['income_to_loan_ratio'] = df['income'] / df['loan_amount']


df['emi_estimate'] = df['loan_amount'] / df['term']


df['emi_income_ratio'] = df['emi_estimate'] / df['income']
```
